```python
import os
import pickle
```

```python
def positional_index(Processed_files):
  positional_index={}
  for file in os.listdir(Processed_files):
    file_path=os.path.join(Processed_files,file)
    with open(file_path,'r',encoding='utf-8') as processed_file:
      words=processed_file.read().split()
      for position,word in enumerate(words):
        if word not in positional_index:
          positional_index[word]={}
        if file not in positional_index[word]:
          positional_index[word][file]=[]
        positional_index[word][file].append(position)
  return positional_index
positional_i=positional_index("Processed_files")

with open("positional_i.pickle",'wb') as f:
  pickle.dump(positional_i,f,protocol=pickle.HIGHEST_PROTOCOL)

with open("positional_i.pickle",'rb') as f:
  loaded_index=pickle.load(f)
  print(loaded_index)
```
